```python
def triangleFinite(w,m1,m2,m3):
    if w^m1.is_one() or w^m2.is_one() or w^m3.is_one():
        return True
    else:
        return False
```

```python
#checks is r1 seperates r2 from the identity
def seperateFromOne(r1,r2,m1,m2,m3):
    r=r1*r2
    if triangleFinite(r,m1,m2,m3):
        return False
    else if (r).length<(r2).length():
        return True
    else:
        return False
```

```python
def wallsOfOne(G,m1,m2,m3):
    fin=False
    parallel=False
    n=1
    s=G.simple_reflections()
    H=list(s)
    while fin==False:
        #if we find any new walls we stop
        fin=True
        #looking at walls at distance n from the identity
        for g in list(G.elements_of_length(n)):
            #got 3 reflections around any point
            for t in s:
                #checking if it is a new one
                for r in H:
                    #check if parallel to a wall we already have, if not prallel to all
                    if not triangleFinite(g*t*(g^-1)*r,m1,m2,m3):
                        parallel=True
                    if ((g^-1)*r).length()==0:
                        parallel=True
                if parallel==False:
                    H.append(g*t)
                    fin=False
        print(n)
        n=n+1
    return H
                
```
